```python
user_br_df = br_df.pivot_table(index='userId', columns='title', values='rating', 
               aggfunc='sum')
 
```

```python
user_br_df
```

```python
user_br_df.index = br_df.userId.unique()
```

```python
user_br_df
```

```python
#Impute those NaNs with 0 values
user_br_df.fillna(0, inplace=True)
```

```python
user_br_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_br_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = br_df.userId.unique()
user_sim_df.columns = br_df.userId.unique()
```
